```python
import pandas as pd
```

```python
business = pd.read_json('meta-Georgia.json', lines=True)
reviews = pd.read_json('review-Georgia.json', lines=True)
```

```python
business
```

```python
reviews
```

```python
restaurants_df = business[business['category'].apply(lambda x: 'Restaurant' in x if x is not None else False)]
```

```python
restaurant_reviews = pd.merge(reviews, restaurants_df, on='gmap_id', how='inner')
```

```python
atlanta_min_lat, atlanta_max_lat = 33.6478, 33.8868
atlanta_min_long, atlanta_max_long = -84.5513, -84.2897
# atlanta_min_lat, atlanta_max_lat = 33.6600, 33.8500
# atlanta_min_long, atlanta_max_long = -84.5000, -84.3300

atlanta_restaurants = restaurant_reviews[(restaurant_reviews['latitude'].between(atlanta_min_lat, atlanta_max_lat)) & 
                                        (restaurant_reviews['longitude'].between(atlanta_min_long, atlanta_max_long))]
```

```python
atlanta_restaurant_reviews = atlanta_restaurants.drop_duplicates(subset=['user_id', 'time', 'gmap_id'], keep='first')
```

```python
atlanta_restaurant_reviews
```

```python
atlanta_restaurant_reviews = atlanta_restaurant_reviews[atlanta_restaurant_reviews['state'] != 'Permanently closed']
```

```python
atlanta_restaurant_reviews
```

```python
# First, define a function to check if the category list contains 'Shopping Mall' or 'Grocery Store'
def filter_categories(categories):
    if categories is not None:
        return not any(category in ['Shopping mall', 'Grocery Store'] for category in categories)
    return True

# Apply the function to the 'category' column and use boolean indexing to filter the rows
atlanta_restaurant_reviews = atlanta_restaurant_reviews[atlanta_restaurant_reviews['category'].apply(filter_categories)]
```

```python
atlanta_restaurant_reviews.to_csv('filtered_atlanta_restaurant_reviews.csv', index=False)
```
